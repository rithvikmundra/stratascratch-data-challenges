```python
df_join_2013 = pd.merge(df_males_2013,df_females_2013, on="name", how="inner").sort_values(["count_x","count_y"], ascending=False)
df_join_2013["total_sum"] = df_join_2013["count_x"] + df_join_2013["count_y"]
df_join_2013
```

```python
df_join_1945 = pd.merge(df_males_1945,df_females_1945, on="name", how="inner").sort_values(["count_x","count_y"], ascending=False)
df_join_1945["total_sum"] = df_join_1945["count_x"] + df_join_1945["count_y"]
df_join_1945
```

### Q4 - Of the names represented in the data, find the name that has had the largest percentage increase in popularity since 1980. Largest decrease?

```python
df_1980 = df[(df['year'] == 1980)].groupby("name")["count"].sum().reset_index().sort_values("count", ascending=False)
df_1980["total"] = df_1980["count"].sum()
df_1980["popularity_pct"] = (df_1980["count"]*100/df_1980["total"]).round(2)
df_1980.sort_values("popularity_pct", ascending=False)
```

```python
df_2021 = df[(df['year'] == 2021)].groupby("name")["count"].sum().reset_index().sort_values("count", ascending=False)
df_2021["total"] = df_2021["count"].sum()
df_2021["popularity_pct"] = (df_2021["count"]*100/df_2021["total"]).round(2)
df_2021.sort_values("popularity_pct", ascending=False)
```

```python
df_1980_2021_joined = pd.merge(df_1980, df_2021, on="name", how="inner")
df_1980_2021_joined
```

```python
df_1980_2021_joined['diff'] = df_1980_2021_joined["popularity_pct_y"] -  df_1980_2021_joined["popularity_pct_x"]
df_1980_2021_joined.sort_values("diff", ascending=False)
```

#### From 1980 through 2021, Liam is the name which gained the highest popularity 

```python
df_1980_2021_joined.sort_values("diff", ascending=True)
```

#### From 1980 through 2021, "Michael" is the name with highest decrease in popularity 

### Q5 - Can you identify names that may have had an even larger increase or decrease in popularity?

```python
df_1910_2021_joined = pd.merge(df_1910, df_2021, on="name", how="inner")
df_1910_2021_joined
```

```python
df_1910_2021_joined['diff'] = df_1910_2021_joined["popularity_pct_y"] -  df_1910_2021_joined["popularity_pct_x"]
df_1910_2021_joined.sort_values("diff", ascending=False)
```
