### Scenario1

Option 1: $50 for each driver that is online at least 8 hours, accepts 90% of requests, completes 10 trips, and has a rating of 4.7 or better during the time frame;

Option 2: $4/trip for all drivers who complete 12 trips, and have a 4.7 or better rating.


#### Q1 - How much would the total bonus payout be with Option 1?

```python
df_option1 = data_df[(data_df["Accept Rate"].str.split('%').str[0].astype(int)>=90) & (data_df["Supply Hours"] >= 8) & (data_df["Rating"] >= 4.7)]
df_option1
```

```python
len(df_option1) * 50
```

#### Q2 - How much would the total bonus payout be with Option 2?

```python
df_option2 = data_df[(data_df["Trips Completed"] >= 12) & (data_df["Rating"] >= 4.7)]
df_option2
```

```python
sum(df_option2["Trips Completed"] * 4)
```

#### Q3 - How many drivers would qualify for a bonus under Option 1 but not under Option 2?

```python
option1_names_list = list(df_option1["Name"])

c = 0

for each in option1_names_list:
    if each not in list(df_option2["Name"]):
        c+=1
```

### Scenario 2: 

A taxi driver currently generates $200 per day in fares (before expenses), works six days a week, takes three weeks off, and has the following expenses:

Gas - $200 per week

Insurance - $400 per month

Vehicle rent (by the week) - $500

The driver doesn't pay gas and rent expenses on off weeks.

Now, let's assume that the same driver would buy a Town Car and partner with Uber. If he does, his gas expenses would go up by 5%, his insurance expense would decrease by 20%, and he would no longer be renting a vehicle. However, he would need to buy a car. The driver would still take three weeks off per year.

Given Scenario 2, provide answers to the questions below:

##### Q5 How much money (after expenses) does the taxi driver make per year without partnering with Uber?

##### Q6 You are convincing the same driver above to buy a Town Car and partner with Uber. Assuming the new car is 40,000 USD, how much would the driver's gross fares need to increase per week to fully pay for the car in year 1 and maintain the same yearly profit margin as before?

```python
gas = 200
insurance = 400
rent = 500
n_weeks = 52
off_weeks = 3
n_days_work=6

total_revenue = (n_weeks-off_weeks)*200*n_days_work
total_revenue
```

```python
total_expenses = (gas + rent) * (n_weeks - off_weeks) + insurance*12
total_expenses
```

```python
total_profit = total_revenue - total_expenses
total_profit
```

```python
margin = total_profit*100/total_revenue
margin
```

```python
new_expenses = (gas*1.05)*(n_weeks - off_weeks) +insurance*0.8*12  + 40000 #+total_profit
new_expenses
```

```python
new_margin = (total_revenue-new_expenses)*100/total_revenue
new_margin
```

```python
total_revenue * ((margin-new_margin)/100) / 49/6
```
